```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
file_path = '../data/sensor_data.csv'
if os.path.exists(file_path):
    print("File exists")
else:
    print("File does not exist")
data = pd.read_csv('../data/sensor_data.csv')

data.head()

```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.subplots_adjust(hspace=0.5)
```

```python
plt.subplot(2, 3, 1)
sns.distplot(data['temp'], bins=20)
plt.title('Temperature Distribution')
```

```python
plt.subplot(2, 3, 2)
sns.distplot(data['vibration'], bins=20)
plt.title('Vibration Distribution')
```

```python
plt.subplot(2, 3, 3)
sns.distplot(data['pressure'], bins=20)
plt.title('Pressure Distribution')
```

```python
plt.subplot(2, 3, 4)
sns.distplot(data['rpm'], bins=20)
plt.title('RPM Distribution')
```

```python
plt.subplot(2, 3, 5)
sns.distplot(data['load'], bins=20)
plt.title('Load Distribution')
```

```python
plt.show()
```
